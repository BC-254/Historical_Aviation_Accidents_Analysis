## **Data Preparation**

Before any meaningful insights can be drawn, the dataset must be free from any significant errors or inconsistencies. As a result, the following data preparation steps were in order:

### 1. Removing Duplicates

```python
# Checking if there are any duplicate entries
df_raw.duplicated().sum()
print('There were', df_raw.duplicated().sum(), 'duplicate entries found.')
```

```python
# Removing the duplicate entries
df_raw.drop_duplicates(inplace=True)
df_raw
```

### 2. Dealing with the errors in Date column

Given the date column contains inconsistencies in the data, I decided to extract the month from the date column and do away with the days as the specific day of accident may not be necessary for my analysis

```python
# Method: Extract the 3-letter month code using Regex by looking for 3 letters between two hyphens (e.g., the "MAY" in "??-MAY-1936")
df_raw['Month_extract'] = df_raw['date'].str.extract(r'-([a-zA-Z]{3})-')

# Standardize: Convert to uppercase to be safe (e.g., 'May' -> 'MAY')
df_raw['Month_extract'] = df_raw['Month_extract'].str.upper()

# Mapping the 3-letter codes to Full Names for better readability
month_map = {
    'JAN': 'January', 'FEB': 'February', 'MAR': 'March', 'APR': 'April',
    'MAY': 'May', 'JUN': 'June', 'JUL': 'July', 'AUG': 'August',
    'SEP': 'September', 'OCT': 'October', 'NOV': 'November', 'DEC': 'December'
}
df_raw['Month'] = df_raw['Month_extract'].map(month_map)

# Check the results
df_raw
```

Now, to drop the 'date' column and the temporary 'Month_extract' column

```python
# Dropping the date column as we have extracted the month information
# Also dropping the Month_extract column as it's no longer needed
df_raw.drop(columns=['date', 'Month_extract'], inplace=True) 
df_raw
```

### 3. Dealing with inconsistencies in the data values

```python
# Changing the year column to datetime format
df_raw['year'] = pd.to_datetime(df_raw['year'], errors='coerce')

# Replacing 'unknown' in the year column with NaT (Not a Time)
df_raw['year'].replace('unknown', pd.NaT, inplace=True)
df_raw

# Replace '?' and 'Unknown country' with a single standard 'Unknown' category in the country column
df_raw['country'] = df_raw['country'].replace(['?', 'Unknown country'], 'Unknown')

# Mapping the broken and invalid text to the correct names of the countries
replacements = {
    'CuraÃ§ao': 'Curacao',
    'Saint BarthÃ©lemy': 'Saint Barthelemy'
}
df_raw['country'] = df_raw['country'].replace(replacements)
```

### 4. Checking and dealing with missing values

The following fields had missing values:

```python
df_raw.isnull().sum()
```

Since my analysis will be heavily time-based, the 'year' column has some null values which cannot be imputed. Instead, dropping these values will be the feasible way to deal with the 61 missing entries.

```python
# Dropping the missing values in the 'year' column
df_raw.dropna(subset=['year'], inplace=True)
```

```python
# Replacing the null values of the 'fatalities' column with the mode 
df_raw['fatalities'] = df_raw['fatalities'].fillna(df_raw['fatalities'].mode()[0])

# Replacing the null values of the 'registration' column with 'unknown'
df_raw['registration'] = df_raw['registration'].fillna('unknown')

# Replacing the null values of the 'operator' column with 'unknown'
df_raw['operator'] = df_raw['operator'].fillna('unknown')

# Replacing the null values of the 'location' column with 'unknown'
df_raw['location'] = df_raw['location'].fillna('unknown')

# Replacing the null values of the 'Month' column with 'Unknown'
df_raw['Month'] = df_raw['Month'].fillna('Unknown')

# Final check for any remaining missing values
df_raw.isnull().sum()
```
